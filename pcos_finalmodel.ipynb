```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import joblib

df = pd.read_csv(r"C:\Users\Janhavi\Desktop\pcos_data (2).csv")

X = df[["Age", "BMI", "Menstrual Regularity", "Acne Severity", "Stress Levels"]]
y = df["Risk"]

model = LogisticRegression(max_iter=1000)
model.fit(X, y)

joblib.dump(model, "pcos_final.pkl")
print("Model saved as pcos_final.pkl")
```
